# get2work
> The Python library that makes you actually work.

get2work is a CLI tool that gamifies your coding productivity. It celebrates your commits, roasts your git history, and shames you when you're slacking.

**What you'll learn in this tutorial:**
- How to install get2work
- How to set up git hooks
- How to use all the commands
- How to configure AI roasts


Every commit will be celebrated

## Set up in your project
Before get2work can celebrate your commits, you need to install the git hook once per project.
This makes get2work automatically trigger every time you run `git commit` — no manual commands needed.

image.png

````python
# Run this command in YOUR terminal (not here in Colab)
# inside the folder of the project you want to track

# get2work --help

# cd your-project
# get2work install

# From that moment, every git commit triggers get2work automatically
print("Run 'get2work install' in your project folder in your terminal!")
````

## Background watcher
Want get2work running 24/7 without thinking about it?
`get2work start` runs in the background watching for commits, checking for distractions every 2 minutes, and shaming you every 30 minutes if you haven't committed. If it finds you in any application on the list of distracting apps (Spotify, Netflix, YouTube, etc.) it shows a pop-up that takes you back to work

````python
# Run this in a separate terminal (not here in Colab)
# It watches your repo 24/7

# get2work start

print("Run 'get2work start' in a separate terminal to keep it watching!")
````

## Level system
8 levels from Hello World Survivor to Linus Would Be Proud

| Level | Name | Commits needed |
|---|---|---|
| 1 | Hello World Survivor | 0 |
| 2 | Tutorial Finisher | 10 |
| 3 | Functional but Confused | 30 |
| 4 | It Works Don't Touch It | 60 |
| 5 | Commits with Confidence (wrongly) | 100 |
| 6 | git push --force and it worked | 200 |
| 7 | The Last Line of Defense | 400 |
| 8 | Linus Would Be Proud | 700 |

Every commit moves you closer to the next level

## Check your status
See your current level, streak and stats

````python
# Check your level and stats
!get2work status
````

## Shame check
No AI involved here — this one is pure math.
Did you commit today? Yes → you're safe.
No → get2work checks how many hours since your last commit
and escalates the message accordingly.
Under 4 hours: gentle nudge. Over 8 hours: full intervention. 💀

````python
!get2work shame
````

## Distraction detector
get2work watches your running processes in the background.
If it catches you with Netflix, Spotify, Discord, TikTok, or any other
non-work app open — it shames you in the terminal AND sends a system
notification so you can't ignore it.

Every time you get caught, it adds to your `distractions_caught` counter
visible in `get2work status`. The shame is permanent and public (to yourself).

````python
# Check if you're slacking right now
!get2work distraction
````

## Custom sounds
Replace default meme sounds with your own
